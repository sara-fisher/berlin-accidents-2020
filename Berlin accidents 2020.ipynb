```python
df = pd.read_csv('AfSBBB_BE_LOR_Strasse_Strassenverkehrsunfaelle_2020_Datensatz.csv', delimiter=';')
```

```python
df.rename(columns={'OBJECTID': 'ID',
                   'LAND': 'State',
                   'BEZ': 'District',
                   'LOR': 'LOR',
                   'LOR_ab_2021': 'LOR-2021',
                   'UJAHR': 'Year',
                   'UMONAT': 'Month',
                   'USTUNDE': 'Hour',
                   'UWOCHENTAG': 'Weekday',
                   'UKATEGORIE': 'Category',
                   'UART': 'Kind',
                   'UTYP1': 'Type',
                   'ULICHTVERH': 'Light',
                   'IstRad': 'Bike',
                   'IstPKW': 'Car',
                   'IstFuss': 'Pedestrian',
                   'IstKrad': 'Motorcycle',
                   'IstGkfz': 'Freight Vehicle',
                   'IstSonstige': 'Other',
                   'USTRZUSTAND': 'Street Condition',
                   'LINREFX': 'LINREFX',
                   'LINREFY': 'LINREFY',
                   'XGCSWGS84': 'Longitude',
                   'YGCSWGS84': 'Latitude'}, inplace = True)
```

```python
df = df.drop('State', axis=1)
```

```python
df.columns
```

```python
df[['LINREFX', 'LINREFY', 'Longitude', 'Latitude']].head()
```

```python
df['Longitude'] = df['Longitude'].apply(lambda x: float(x.replace(',', '.')))
df['Latitude'] = df['Latitude'].apply(lambda x: float(x.replace(',', '.')))
df['LINREFX'] = df['LINREFX'].apply(lambda x: float(x.replace(',', '.')))
df['LINREFY'] = df['LINREFY'].apply(lambda x: float(x.replace(',', '.')))
```

```python
df.loc[df['LOR-2021'].isna()]
```

```python
wd_map = {1 : 'Sunday', 
          2 : 'Monday',
          3 : 'Tuesday',
          4 : 'Wednesday',
          5 : 'Thursday',
          6 : 'Friday',
          7 : 'Saturday'}
df['Weekday'] = df['Weekday'].map(wd_map)
```

```python
sev_map = {1 : 'Deadly', 
          2 : 'Severe',
          3 : 'Light'}
df['Category'] = df['Category'].map(sev_map)
```

```python
light_map = {0: 'Day',
             1: 'Dusk/Dawn',
             2: 'Dark'}
df['Light'] = df['Light'].map(light_map)
```

```python
kind_map = {1: 'Collision with approaching/stopping/stationary vehicle',
            2: 'Collision with vehicle ahead/waiting vehicle',
            3: 'Collision with side vehicle driving in same direction',
            4: 'Collision with oncoming vehicle',
            5: 'Collision with turning/crossing vehicle',
            6: 'Collision between vehicle and pedestrian',
            7: 'Impact with road obstacle',
            8: 'Left roadway to the right',
            9: 'Left roadway to the left',
            0: 'Other'}
df['Kind'] = df['Kind'].map(kind_map)
```

```python
type_map = {1: 'Driving',
            2: 'Turning',
            3: 'Turning in/Crossing',
            4: 'Speeding',
            5: 'Idling traffic',
            6: 'Parallel traffic',
            7: 'Other'}
df['Type'] = df['Type'].map(type_map)
```

```python
bike_map = {0: 'No bikes involved',
            1: 'Bike(s) involved'}
df['Bike'] = df['Bike'].map(bike_map)
```

```python
car_map = {0: 'No cars involved',
           1: 'Car(s) involved'}
df['Car'] = df['Car'].map(car_map)
```

```python
ped_map = {0: 'No pedestrians involved',
           1: 'Pedestrian(s) involved'}
df['Pedestrian'] = df['Pedestrian'].map(ped_map)
```

```python
moto_map = {0: 'No motorcycles involved',
            1: 'Motorcycle(s) involved'}
df['Motorcycle'] = df['Motorcycle'].map(moto_map)
```

```python
freight_map = {0: 'No freight vehicles involved',
               1: 'Freight vehicle(s) involved'}
df['Freight Vehicle'] = df['Freight Vehicle'].map(freight_map)
```

```python
other_map = {0: 'No other means of transport involved',
             1: 'Other means of transport involved'}
df['Other'] = df['Other'].map(other_map)
```

```python
street_map = {0: 'Dry',
              1: 'Wet/Damp/Slick',
              2: 'Icy'}
df['Street Condition'] = df['Street Condition'].map(street_map)
```

```python
assert df['ID'].nunique()==df['ID'].shape[0]
```
